```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import TruncatedSVD
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd

# Strip leading/trailing spaces from column names
df6.columns = df6.columns.str.strip()

# Define high-cardinality categorical columns (stripped names)
high_card_cols = ['post_impacts', 'favorite_workout', 'biggest_challenge']

# Step 1: Create OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore', sparse=True)

# Step 2: Dimensionality reduction using TruncatedSVD
svd = TruncatedSVD(n_components=5, random_state=42)

# Step 3: Build pipeline
reduction_pipeline = Pipeline([
    ('ohe', ohe),
    ('svd', svd)
])

# Step 4: Fit-transform each column and save new features
for col in high_card_cols:
    reduced_features = reduction_pipeline.fit_transform(df6[[col]])
    
    # Create new column names
    reduced_df = pd.DataFrame(reduced_features, 
                              columns=[f"{col}_svd_{i}" for i in range(1, 6)],
                              index=df6.index)
    
    # Concatenate to original dataframe
    df6 = pd.concat([df6.drop(columns=[col]), reduced_df], axis=1)

# Optional: Check updated shape
print(df6.shape)
```

```python
# -----------------------------
# ABLATION STUDY (SELF-CONTAINED)
# -----------------------------
import numpy as np
import pandas as pd

from sklearn.model_selection import StratifiedKFold, cross_validate, cross_val_predict
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import label_binarize
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, mutual_info_classif

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

cv = 5
skf = StratifiedKFold(n_splits=cv, shuffle=True, random_state=42)

# ✅ Define models here so "models" is always available
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Linear SVC': LinearSVC(max_iter=10000, random_state=42),
    'k-Nearest Neighbors': KNeighborsClassifier(),
    'Naïve Bayes': GaussianNB(),
    'Random Forest': RandomForestClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    'LightGBM': LGBMClassifier(random_state=42),
    'CatBoost': CatBoostClassifier(verbose=0, random_state=42)
}

def make_ablation_pipeline(clf, use_smote=False, use_fs=False, k_best=30):
    """
    Scale -> (Feature Selection) -> (SMOTE) -> Classifier
    SMOTE is inside CV folds (no leakage) via ImbPipeline.
    """
    steps = [("scaler", StandardScaler())]

    if use_fs:
        steps.append(("fs", SelectKBest(score_func=mutual_info_classif, k=k_best)))

    if use_smote:
        steps.append(("smote", SMOTE(random_state=42)))

    steps.append(("clf", clf))
    return ImbPipeline(steps)

def cv_metrics(estimator, X, y):
    scoring = {
        "acc": "accuracy",
        "prec": "precision_macro",
        "rec": "recall_macro",
        "f1": "f1_macro",
    }

    out = cross_validate(estimator, X, y, cv=skf, scoring=scoring, n_jobs=-1)

    # AUC-ROC (macro, OVO) using out-of-fold scores
    auc_val = np.nan
    try:
        classes = np.unique(y)
        y_bin = label_binarize(y, classes=classes)

        if hasattr(estimator, "predict_proba"):
            y_score = cross_val_predict(estimator, X, y, cv=skf, method="predict_proba", n_jobs=-1)
        elif hasattr(estimator, "decision_function"):
            y_score = cross_val_predict(estimator, X, y, cv=skf, method="decision_function", n_jobs=-1)
        else:
            y_score = None

        if y_score is not None:
            auc_val = roc_auc_score(y_bin, y_score, multi_class="ovo", average="macro")
    except Exception:
        auc_val = np.nan

    return {
        "Accuracy (mean ± std)": f"{out['test_acc'].mean():.4f} ± {out['test_acc'].std():.4f}",
        "Precision (mean ± std)": f"{out['test_prec'].mean():.4f} ± {out['test_prec'].std():.4f}",
        "Recall (mean ± std)": f"{out['test_rec'].mean():.4f} ± {out['test_rec'].std():.4f}",
        "F1 (mean ± std)": f"{out['test_f1'].mean():.4f} ± {out['test_f1'].std():.4f}",
        "AUC-ROC (OOF)": (f"{auc_val:.4f}" if not np.isnan(auc_val) else "nan"),
    }

# ---- Define ablation settings ----
k_best = min(30, X_train.shape[1])  # safe if you have <30 features

ablations = [
    ("Baseline (Scale only)", dict(use_smote=False, use_fs=False)),
    ("+ SMOTE",               dict(use_smote=True,  use_fs=False)),
    (f"+ FS (k={k_best})",    dict(use_smote=False, use_fs=True,  k_best=k_best)),
    (f"+ FS (k={k_best}) + SMOTE", dict(use_smote=True, use_fs=True, k_best=k_best)),
]

# ✅ IMPORTANT: use ORIGINAL training set (no resampling pre-done)
# X_train, y_train should be your raw train split
ablation_rows = []
for ab_name, ab_cfg in ablations:
    for model_name, clf in models.items():
        pipe = make_ablation_pipeline(clf, **ab_cfg)
        m = cv_metrics(pipe, X_train, y_train)
        ablation_rows.append({"Ablation": ab_name, "Model": model_name, **m})

ablation_df = pd.DataFrame(ablation_rows)

print("\n📌 Ablation Study Results (CV on training set):")
print(ablation_df.to_string(index=False))

# Optional save
# ablation_df.to_csv("ablation_results.csv", index=False)

# Optional LaTeX dump
# print(ablation_df.to_latex(index=False, escape=False))
```

```python
# -----------------------------
# ABLATION: SMOTE + FEATURE REDUCTION (21 -> 15 -> 10)
# -----------------------------
import numpy as np
import pandas as pd

from sklearn.model_selection import StratifiedKFold, cross_validate, cross_val_predict
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import label_binarize

from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

# --- (use your same model imports if already loaded) ---
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

cv = 5
skf = StratifiedKFold(n_splits=cv, shuffle=True, random_state=42)

# Define models (edit if you want fewer models for faster runtime)
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Linear SVC': LinearSVC(max_iter=10000, random_state=42),
    'k-NN': KNeighborsClassifier(),
    'Naïve Bayes': GaussianNB(),
    'Random Forest': RandomForestClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    'LightGBM': LGBMClassifier(random_state=42),
    'CatBoost': CatBoostClassifier(verbose=0, random_state=42),
}

def make_pipeline(clf, use_fs=False, k_best=None, use_smote=False):
    """
    Scale -> (FS) -> (SMOTE) -> Classifier
    FS + SMOTE happen inside CV folds via ImbPipeline => no data leakage.
    """
    steps = [("scaler", StandardScaler())]

    if use_fs:
        steps.append(("fs", SelectKBest(score_func=mutual_info_classif, k=k_best)))

    if use_smote:
        steps.append(("smote", SMOTE(random_state=42)))

    steps.append(("clf", clf))
    return ImbPipeline(steps)

def compute_metrics(estimator, X, y):
    scoring = {
        "acc": "accuracy",
        "prec": "precision_macro",
        "rec": "recall_macro",
        "f1": "f1_macro",
    }
    out = cross_validate(estimator, X, y, cv=skf, scoring=scoring, n_jobs=-1)

    # OOF AUC-ROC (macro, OVO) if possible
    auc_val = np.nan
    try:
        classes = np.unique(y)
        y_bin = label_binarize(y, classes=classes)

        if hasattr(estimator, "predict_proba"):
            y_score = cross_val_predict(estimator, X, y, cv=skf, method="predict_proba", n_jobs=-1)
        elif hasattr(estimator, "decision_function"):
            y_score = cross_val_predict(estimator, X, y, cv=skf, method="decision_function", n_jobs=-1)
        else:
            y_score = None

        if y_score is not None:
            auc_val = roc_auc_score(y_bin, y_score, multi_class="ovo", average="macro")
    except Exception:
        auc_val = np.nan

    return {
        "Accuracy (mean ± std)": f"{out['test_acc'].mean():.4f} ± {out['test_acc'].std():.4f}",
        "Precision (mean ± std)": f"{out['test_prec'].mean():.4f} ± {out['test_prec'].std():.4f}",
        "Recall (mean ± std)": f"{out['test_rec'].mean():.4f} ± {out['test_rec'].std():.4f}",
        "F1 (mean ± std)": f"{out['test_f1'].mean():.4f} ± {out['test_f1'].std():.4f}",
        "AUC-ROC (OOF)": (f"{auc_val:.4f}" if not np.isnan(auc_val) else "nan"),
    }

# ---- Feature counts you want to test (21 -> 15 -> 10) ----
# Auto-safety: don't request more than you actually have
p = X_train.shape[1]
k_list = [k for k in [21, 15, 10] if k <= p]

# ---- Ablation grid ----
# Baseline (no FS), FS-only for each k, and both variants with SMOTE
ablation_specs = [("Baseline (Scale only)", dict(use_fs=False, k_best=None, use_smote=False)),
                  ("+ SMOTE",               dict(use_fs=False, k_best=None, use_smote=True))]

for k in k_list:
    ablation_specs.append((f"+ FS (k={k})", dict(use_fs=True, k_best=k, use_smote=False)))
    ablation_specs.append((f"+ FS (k={k}) + SMOTE", dict(use_fs=True, k_best=k, use_smote=True)))

# ---- Run ablation study ----
rows = []
for ab_name, cfg in ablation_specs:
    for model_name, clf in models.items():
        pipe = make_pipeline(clf, **cfg)
        m = compute_metrics(pipe, X_train, y_train)  # ✅ raw X_train, y_train
        rows.append({
            "Ablation": ab_name,
            "Model": model_name,
            "k (if FS)": (cfg["k_best"] if cfg["use_fs"] else "-"),
            "SMOTE": ("Yes" if cfg["use_smote"] else "No"),
            **m
        })

ablation_df = pd.DataFrame(rows)

print("\n📌 Ablation Study Results (5-fold CV on training set):")
print(ablation_df.to_string(index=False))

# Optional: save
# ablation_df.to_csv("ablation_fs_smote_results.csv", index=False)

# Optional: quick “best per ablation” view (by AUC then F1)
# (AUC is string, convert if you want)
# ablation_df["AUC_num"] = pd.to_numeric(ablation_df["AUC-ROC (OOF)"], errors="coerce")
# best_df = (ablation_df.sort_values(["Ablation","AUC_num"], ascending=[True,False])
#                      .groupby("Ablation", as_index=False).head(1))
# print("\n🏆 Best model per ablation (by AUC):")
# print(best_df[["Ablation","Model","Accuracy (mean ± std)","F1 (mean ± std)","AUC-ROC (OOF)"]].to_string(index=False))
```
