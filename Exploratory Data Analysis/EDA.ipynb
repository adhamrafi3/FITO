```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from textwrap import fill

# --- Styling ---
plt.rcParams['font.family'] = 'Times New Roman'
TITLE_SIZE = 20
AXIS_LABEL_SIZE = 18
TICK_LABEL_SIZE = 14
ANNOT_SIZE = 12

# Desired left→right x labels
x_order = ["Minimal to Mild", "Moderate", "Substantial to Exceptional"]
health_map = {0: x_order[0], 1: x_order[1], 2: x_order[2]}

df4 = df4.copy()
if pd.api.types.is_numeric_dtype(df4["health_improvement"]):
    df4["health_label"] = df4["health_improvement"].map(health_map)
else:
    df4["health_label"] = df4["health_improvement"]

# Desired top→bottom y labels
y_order = [
    "No, not supportive at all",
    "Neutral, neither supportive nor unsupportive",
    "Yes, supportive"
]

df4["health_label"] = pd.Categorical(df4["health_label"], categories=x_order, ordered=True)
df4["family_reaction"] = pd.Categorical(df4["family_reaction"], categories=y_order, ordered=True)

# Crosstab with full grid
ct = pd.crosstab(df4["family_reaction"], df4["health_label"])
ct = ct.reindex(index=y_order, columns=x_order, fill_value=0).astype(int)

# ---- Plot ----
fig, ax = plt.subplots(figsize=(10, 6))  # a bit wider helps
hm = sns.heatmap(
    ct, annot=True, fmt='d', cmap='YlGnBu', cbar=True,
    annot_kws={'size': ANNOT_SIZE}, ax=ax
)

ax.set_title("Counts of Family Reaction vs Health Improvement", fontsize=TITLE_SIZE)
ax.set_xlabel("Health Improvement", fontsize=AXIS_LABEL_SIZE)
ax.set_ylabel("Family Reaction", fontsize=AXIS_LABEL_SIZE)

# Wrap long tick labels so they don't collide
wrapped_x = [fill(lbl, 16) for lbl in ct.columns]  # wrap at ~16 chars
wrapped_y = [fill(lbl, 24) for lbl in ct.index]
ax.set_xticklabels(wrapped_x, fontsize=TICK_LABEL_SIZE, rotation=0, ha='center')
ax.set_yticklabels(wrapped_y, fontsize=TICK_LABEL_SIZE, rotation=0)

# Colorbar tick size
hm.collections[0].colorbar.ax.tick_params(labelsize=TICK_LABEL_SIZE)

# Give extra bottom margin for the multi-line x labels
plt.subplots_adjust(bottom=0.22)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import TruncatedSVD
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd

# Strip leading/trailing spaces from column names
df6.columns = df6.columns.str.strip()

# Define high-cardinality categorical columns (stripped names)
high_card_cols = ['post_impacts', 'favorite_workout', 'biggest_challenge']

# Step 1: Create OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore', sparse=True)

# Step 2: Dimensionality reduction using TruncatedSVD
svd = TruncatedSVD(n_components=5, random_state=42)

# Step 3: Build pipeline
reduction_pipeline = Pipeline([
    ('ohe', ohe),
    ('svd', svd)
])

# Step 4: Fit-transform each column and save new features
for col in high_card_cols:
    reduced_features = reduction_pipeline.fit_transform(df6[[col]])
    
    # Create new column names
    reduced_df = pd.DataFrame(reduced_features, 
                              columns=[f"{col}_svd_{i}" for i in range(1, 6)],
                              index=df6.index)
    
    # Concatenate to original dataframe
    df6 = pd.concat([df6.drop(columns=[col]), reduced_df], axis=1)

# Optional: Check updated shape
print(df6.shape)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

# Set font globally
plt.rcParams['font.family'] = 'Times New Roman'

# Separate features and target
X = df7.drop('health_improvement', axis=1)
y = df7['health_improvement']

# Use only numeric data
X = X.select_dtypes(include='number')

# Fit model using RFE
model = RandomForestClassifier(random_state=42)
rfe = RFE(model, n_features_to_select=10)
rfe.fit(X, y)

# Create results DataFrame
rfe_results = pd.DataFrame({
    'Feature': X.columns,
    'Ranking': rfe.ranking_,
    'Selected': rfe.support_
}).sort_values(by='Ranking')

# Set seaborn theme
sns.set_theme(style='whitegrid')

# Define color palette: green if selected, light grey otherwise
bar_colors = ['#2ecc71' if selected else '#d3d3d3' for selected in rfe_results['Selected']]

# Plot
plt.figure(figsize=(14, 7))
ax = sns.barplot(
    data=rfe_results,
    x='Feature',
    y='Ranking',
    palette=bar_colors
)

# Annotate rankings on top of bars
for i, row in enumerate(rfe_results.itertuples()):
    ax.text(
        i, 
        row.Ranking + 0.1, 
        str(row.Ranking), 
        ha='center', 
        va='bottom',
        fontsize=9,
        color='black'
    )

# Style adjustments
plt.title('RFE Feature Rankings', fontsize=16, weight='bold')
plt.xlabel('Feature', fontsize=12)
plt.ylabel('Ranking', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', linewidth=0.5)

plt.show()
```
