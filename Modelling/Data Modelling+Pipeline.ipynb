```python
from sklearn.preprocessing import LabelEncoder

# All intended categorical columns to encode
columns_to_encode = [
    'gender', 'current_fitness_lvl', 'workout_companion', 'primary_goal', 'biggest_mental_shift', 
    'impact_on_daylife', 'diet_sleep_impact', 'fitness_tracker', 'post_impacts',
    'sleeptime', 'follow_dietroutine', 'hydration_attention', 'favorite_workout ',
    'music_impact', 'recovery_methods', 'additional_impacts', 'family_reaction', 'biggest_challenge'
]

# Only encode columns that exist in df5
for column in columns_to_encode:
    if column in df5.columns:
        le = LabelEncoder()
        df5[column] = le.fit_transform(df5[column].astype(str))

# Clean and map the target variable
df5['health_improvement'] = df5['health_improvement'].str.strip().str.lower()

health_improvement_mapping_updated = {
    'minimal to mild': 0,
    'moderate': 1,
    'substantial to exceptional': 2
}

df5['health_improvement'] = df5['health_improvement'].map(health_improvement_mapping_updated)

# Display a few rows to confirm
print(df5.head())
```

```python
#Dimentionality Reduction 
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import TruncatedSVD
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd

# Strip leading/trailing spaces from column names
df6.columns = df6.columns.str.strip()

# Define high-cardinality categorical columns (stripped names)
high_card_cols = ['post_impacts', 'favorite_workout', 'biggest_challenge']

# Step 1: Create OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore', sparse=True)

# Step 2: Dimensionality reduction using TruncatedSVD
svd = TruncatedSVD(n_components=5, random_state=42)

# Step 3: Build pipeline
reduction_pipeline = Pipeline([
    ('ohe', ohe),
    ('svd', svd)
])

# Step 4: Fit-transform each column and save new features
for col in high_card_cols:
    reduced_features = reduction_pipeline.fit_transform(df6[[col]])
    
    # Create new column names
    reduced_df = pd.DataFrame(reduced_features, 
                              columns=[f"{col}_svd_{i}" for i in range(1, 6)],
                              index=df6.index)
    
    # Concatenate to original dataframe
    df6 = pd.concat([df6.drop(columns=[col]), reduced_df], axis=1)

# Optional: Check updated shape
print(df6.shape)
```

```python
# Feature Interaction: Creating new features

# 1. Add follow_dietroutine and hydration_attention to create a discipline score
df6['discipline_score'] = df6['follow_dietroutine'] + df6['hydration_attention']

# 2. Subtract music_impact from scale_feelings to estimate mental resilience
df6['mental_resilience'] = df6['scale_feelings'] - df6['music_impact']

# 3. Multiply exercise_timeline_weekly and current_fitness_lvl to estimate lifestyle commitment
df6['lifestyle_commitment'] = df6['exercise_timeline_weekly'] * df6['current_fitness_lvl']

# Preview the new features
print(df6[['discipline_score', 'mental_resilience', 'lifestyle_commitment']].head())
```

<h1>Train-Test Split</h1>

```python
from sklearn.model_selection import train_test_split

# Separate features and target
X = df8.drop('health_improvement', axis=1)
y = df8['health_improvement']

# Train-test split (15% test size with stratification)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.15, stratify=y, random_state=42
)

print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")
```

<h1>Resampling Training_Set with SMOTE </h1>

```python
from imblearn.combine import SMOTETomek

# Initialize SMOTETomek with random_state
smotetomek = SMOTETomek(random_state=42)

# Apply only to training data (avoid leakage!)
X_train_res, y_train_res = smotetomek.fit_resample(X_train, y_train)

print(f"Resampled train shape: {X_train_res.shape}, Resampled target distribution:\n{y_train_res.value_counts()}")
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Set global font to Times New Roman
rcParams['font.family'] = 'Times New Roman'

# Step 1: Show distributions
print("Original training target distribution:\n", y_train.value_counts())
print("\nResampled training target distribution:\n", y_train_res.value_counts())

# Step 2: Visualization
# Prepare counts
original_counts = y_train.value_counts().sort_index()
resampled_counts = y_train_res.value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.bar(original_counts.index - 0.2, original_counts.values, width=0.4, label='Original')
plt.bar(resampled_counts.index + 0.2, resampled_counts.values, width=0.4, label='After SMOTETomek')
plt.xlabel('Class', fontsize=14)
plt.ylabel('Number of Samples', fontsize=14)
plt.title('Class Distribution Before and After SMOTETomek', fontsize=16)
plt.xticks(sorted(y_train.unique()), fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Step 3: Explanation (print out)
print("\n📑 Why apply SMOTETomek only to training data?")
print("Insights")
print(" - The test set should represent real-world class imbalance.")
print(" - Applying resampling to test data could give inflated performance metrics.")
print(" - Therefore, we apply SMOTETomek only to the training set for proper model evaluation.")
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit on training data and transform both train and test sets
X_train_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test)

print("Scaling complete. Ready for model training!")
```

# MODEL BUILDING

```python
from sklearn.model_selection import cross_val_score, cross_val_predict, StratifiedKFold
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, 
    confusion_matrix, roc_curve, auc
)
from sklearn.preprocessing import label_binarize
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import (
    RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier,
    StackingClassifier, VotingClassifier
)
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
import pandas as pd

rcParams['font.family'] = 'Times New Roman'
cv = 5

summary_metrics = []

# Define class names mapping
class_labels = ["Min to Mild", "Moderate", "Sub to Excep"]

def evaluate_model(model, model_name, X_train, y_train, X_test, y_test):
    skf = StratifiedKFold(n_splits=cv, shuffle=True, random_state=42)
    scores = {
        'accuracy': cross_val_score(model, X_train, y_train, cv=skf, scoring='accuracy'),
        'precision': cross_val_score(model, X_train, y_train, cv=skf, scoring='precision_macro'),
        'recall': cross_val_score(model, X_train, y_train, cv=skf, scoring='recall_macro'),
        'f1': cross_val_score(model, X_train, y_train, cv=skf, scoring='f1_macro'),
    }
    try:
        probs = cross_val_predict(model, X_train, y_train, cv=skf, method='predict_proba')
        auc_cv = roc_auc_score(label_binarize(y_train, classes=np.unique(y_train)), probs, multi_class='ovo', average='macro')
        auc_std = 0
    except:
        auc_cv = np.nan
        auc_std = np.nan

    metrics = ['Accuracy', 'Precision', 'Recall', 'F1', 'AUC-ROC']
    means = [scores[m].mean() for m in scores] + [auc_cv]
    stds = [scores[m].std() for m in scores] + [auc_std]

    summary_metrics.append({
        'Model': model_name,
        'Accuracy (mean ± std)': f"{means[0]:.4f} ± {stds[0]:.4f}",
        'Precision (mean ± std)': f"{means[1]:.4f} ± {stds[1]:.4f}",
        'Recall (mean ± std)': f"{means[2]:.4f} ± {stds[2]:.4f}",
        'F1-Score (mean ± std)': f"{means[3]:.4f} ± {stds[3]:.4f}",
        'AUC-ROC (train)': f"{means[4]:.4f}"
    })

    print(f"\n🔎 {model_name} Performance ({cv}-Fold CV):")
    for i, m in enumerate(metrics):
        if not np.isnan(means[i]):
            print(f"{m}: {means[i]:.4f} ± {stds[i]:.4f}")

    plt.figure(figsize=(10,5))
    plt.bar(metrics, means, yerr=stds, capsize=5, color='steelblue')
    plt.ylim(0,1)
    plt.title(f'{model_name} Cross-Validation Performance')
    plt.ylabel('Score')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Confusion Matrix with Class Labels
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=class_labels, yticklabels=class_labels)
    plt.title(f'Confusion Matrix: {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

    # FP and FN Plot
    FP = cm.sum(axis=0) - np.diag(cm)
    FN = cm.sum(axis=1) - np.diag(cm)
    plt.figure(figsize=(6,5))
    plt.bar(['False Positives', 'False Negatives'], [FP.sum(), FN.sum()], color=['red', 'orange'])
    plt.title(f'FP & FN: {model_name}')
    plt.ylabel('Count')
    plt.show()

    # AUC-ROC Curve with Class Labels
    try:
        if hasattr(model, "predict_proba"):
            y_score = model.predict_proba(X_test)
        elif hasattr(model, "decision_function"):
            y_score = model.decision_function(X_test)
        else:
            y_score = None
        if y_score is not None:
            y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
            fpr, tpr, roc_auc = dict(), dict(), dict()
            plt.figure(figsize=(7,5))
            for i in range(y_test_bin.shape[1]):
                fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
                roc_auc[i] = auc(fpr[i], tpr[i])
                plt.plot(fpr[i], tpr[i], label=f'{class_labels[i]} (AUC={roc_auc[i]:.2f})')
            plt.plot([0,1],[0,1],'k--')
            plt.xlabel('False Positive Rate')
            plt.ylabel('True Positive Rate')
            plt.title(f'AUC-ROC Curve: {model_name}')
            plt.legend()
            plt.show()
    except Exception as e:
        print(f"⚠️ AUC-ROC visualization failed: {e}")

# Define core models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Linear SVC': LinearSVC(max_iter=10000, random_state=42),
    'k-Nearest Neighbors': KNeighborsClassifier(),
    'Naïve Bayes': GaussianNB(),
    'Random Forest': RandomForestClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    'LightGBM': LGBMClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'CatBoost': CatBoostClassifier(verbose=0, random_state=42)
}

# Evaluate all models
for name, model in models.items():
    evaluate_model(model, name, X_train_scaled, y_train_res, X_test_scaled, y_test)

# Stacking and Blending
base_estimators = [
    ('rf', RandomForestClassifier(random_state=42)),
    ('xgb', XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)),
    ('lgb', LGBMClassifier(random_state=42))
]
meta_model = RidgeClassifier()
stack_model = StackingClassifier(estimators=base_estimators, final_estimator=meta_model, cv=cv)
blend_model = VotingClassifier(estimators=base_estimators, voting='soft')

evaluate_model(stack_model, 'Stacking', X_train_scaled, y_train_res, X_test_scaled, y_test)
evaluate_model(blend_model, 'Blending', X_train_scaled, y_train_res, X_test_scaled, y_test)

# Summary Table
summary_df = pd.DataFrame(summary_metrics)
print("\n📊 Final Summary Table of Metrics:")
print(summary_df)
```
